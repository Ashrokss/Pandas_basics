# Merging 

```python
week1 = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/RestaurantWeek1Sales.csv")
week2 = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/RestaurantWeek2Sales.csv")
customers = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/RestaurantCustomers.csv")
foods = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/RestaurantFoods.csv")
satisfaction = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/RestaurantWeek1Satisfaction.csv")
```

```python
week1.shape
```

```python
week2.shape
```

```python
week1
```

```python
week2
```

```python
# Concatination 
sales = pd.concat([week1,week2])
```

```python
sales
```

```python
sales.duplicated().sum()
```

```python
week1.merge(week2,how='inner',on = 'Customer ID',suffixes = ('_week1','__week2'))
```

```python
week1.merge(week2,how='outer',on = 'Customer ID',suffixes = ('_week1','__week2'))
```

```python
week1.merge(week2,how='left',on = 'Customer ID',suffixes = ('_week1','__week2'))
```
